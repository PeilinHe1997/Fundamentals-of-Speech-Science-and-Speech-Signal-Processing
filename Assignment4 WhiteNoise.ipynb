```python
import random
input_spec = []
for i in range(0,6000):
    input_spec.append(random.uniform(0, 20))
```

```python
frequency = []
for i in range(0,6000):
    frequency.append(i)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(frequency,input_spec)
plt.ylabel("dB")
```

```python
# response
b1 = -30
a1 = 1/10
b2 = 260/7
a2 = -2/70
result = []
for i in range(0,6000):
    if i<= 300:
        result.append(a1*i+b1)
    elif i> 1300:
        result.append(a2*i+b2)
    else:
        result.append(0)
        
output = []
for i in range(0,6000):
    output1 = result[i]+input_spec[i]
    output.append(output1)
```

```python
plt.plot(frequency,output)
```
